```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Yillarni son sifatida (raqam shaklida) yaratamiz
years = np.array(range(2005, 2025)).reshape(-1, 1)  # 2005–2024

# Maqsad (y) qiymatlarini tayyorlaymiz
uzb_df = df[df['Country Name'] == 'Uzbekistan']
values = uzb_df.loc[:, '2005':'2024'].T.squeeze().values

# Train/Test ma'lumotlarni ajratamiz (70/30)
split_point = int(len(years) * 0.7)
X_train, X_test = years[:split_point], years[split_point:]
y_train, y_test = values[:split_point], values[split_point:]

# Model yaratamiz va o‘qitamiz
model = LinearRegression()
model.fit(X_train, y_train)

print("Model o‘qitildi ✅")
```

```python
# 2024 yilgacha o‘qitilgan model asosida 2025 yilni prognoz qilamiz
y_pred = model.predict(X_test)

# Bashorat qilingan qiymatlarni ko‘ramiz
print("Prognoz qilingan qiymatlar (test to‘plami uchun):")
print(y_pred)
```

```python
print("y_test ichida NaN soni:", np.isnan(y_test).sum())
print("y_pred ichida NaN soni:", np.isnan(y_pred).sum())
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error

# NaN qiymatlarni olib tashlaymiz
mask = ~np.isnan(y_test)

# faqat haqiqiy qiymatlar bilan hisoblaymiz
rmse = np.sqrt(mean_squared_error(y_test[mask], y_pred[mask]))

print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```

```python
from sklearn.metrics import r2_score
import numpy as np

# NaN qiymatlar uchun niqob (mask)
mask = ~np.isnan(y_test)

# faqat haqiqiy (NaN bo‘lmagan) qiymatlar bilan hisoblash
r2 = r2_score(y_test[mask], y_pred[mask])

print('R2:', np.round(r2, 2))
```

```python
import numpy as np
from sklearn.metrics import r2_score

# NaN bo‘lgan joylarni chiqarib tashlaymiz
mask = ~np.isnan(y_test)

r2 = r2_score(y_test[mask], y_pred[mask])
score = model.score(X_test[mask], y_test[mask])

print("r2_score:", r2)
print("model.score:", score)
print("Bir xilmi:", np.isclose(r2, score))
```

```python
# Test to‘plamdagi o‘lchamlarni aniqlaymiz
n, k = X_test.shape

print(f"Namuna soni (n): {n}")
print(f"Belgi soni (k): {k}")
```

```python
import numpy as np

# y_test ni Pandas Series ga o‘tkazamiz
import pandas as pd
y_test = pd.Series(y_test)

# NaN qiymatlarni aniqlaymiz
mask = ~y_test.isna()

# Faqat NaN bo‘lmaganlarni qoldiramiz
X_test_clean = X_test[mask]
y_test_clean = y_test[mask]

# n — namunalar soni, k — belgi soni
n, k = X_test_clean.shape

# Adjusted R² (moslashtirilgan determinatsiya koeffitsienti)
adj_r2 = 1 - (1 - model.score(X_test_clean, y_test_clean)) * ((n - 1) / (n - k - 1))
print(f"Adjusted R²: {adj_r2:.4f}")
```
